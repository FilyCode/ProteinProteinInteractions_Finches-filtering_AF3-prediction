```python
import pyarrow
import pyarrow.parquet as pq
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

input_filename = 'results/finches_interaction_results_chunked.parquet'

df = pq.read_table(input_filename).to_pandas()
print(df.columns)

df_attractive = df[df['interaction_type'] == 'attraction']
df_repulsive = df[df['interaction_type'] == 'repulsion']

print(len(df_attractive))
print(len(df_repulsive))
```

```python
df_na = df[df['interaction_type'] == 'N/A']
print(len(df_na))
na_viruses = df['s1_id'].unique()
print(len(na_viruses))
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='mean_epsilon', y='std_epsilon',
                hue='interaction_type', alpha=0.7, palette='viridis')
plt.title('Mean vs. Std Epsilon by Interaction')
plt.xlabel('Mean Epsilon')
plt.ylabel('Std.dev Epsilon')
plt.legend(title='Interaction', bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.tight_layout() 
plt.show()
```

```python
df_melted = df.melt(id_vars=['s1_id', 's2_id'],
                    value_vars=['mf_s1_s2', 'mf_s2_s1', 'cf_s1_s2', 'cf_s2_s1'],
                    var_name='metric_type',
                    value_name='value')
```

```python
# Overall distribution of each metric
plt.figure(figsize=(15, 5))
ax = sns.histplot(data=df_melted, x='value', hue='metric_type', palette='tab10', alpha=0.6)
plt.title('Overall Distribution of All Four Metrics')
plt.xlabel('Metric Value')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Overall log distribution of each metric
plt.figure(figsize=(15, 5))
ax = sns.histplot(data=df_melted, x='value', hue='metric_type', palette='tab10', alpha=0.6)
plt.yscale('log')
plt.title('Overall Distribution of All Four Metrics')
plt.xlabel('Metric Value')
plt.ylabel('log(Count)')
plt.tight_layout()
plt.show()

# Boxplot comparing the overall distributions of the 4 metrics
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x='metric_type', y='value')
plt.title('Overall Boxplot of All Four Metrics')
plt.xlabel('Metric Type')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

```python
# Aggregate mf_s1_s2 by s2_id to get an average value for each s2_id
# This helps if multiple s1_ids interact with the same s2_id,
# or if you want to see an "overall s2_id trend".
s2_avg_mf_s1_s2 = df.groupby('s2_id')['mf_s1_s2'].mean().reset_index()

# Plot the trend of average mf_s1_s2 across all s2_ids
plt.figure(figsize=(15, 6))
sns.lineplot(data=s2_avg_mf_s1_s2, x='s2_id', y='mf_s1_s2', alpha=0.7)
plt.title('Average mf_s1_s2 across s2_ids (Overall Trend)')
plt.xlabel('s2_id')
plt.ylabel('Average mf_s1_s2')
plt.tight_layout()
plt.show()

# To look at specific s1_id trends across s2_ids (requires further filtering)
# This will still be dense for many s2_ids, but possible if the s1_id has few interactions
selected_s1_id_for_line = 'S1_123' # Choose an s1_id you want to investigate
df_filtered_s1 = df[df['s1_id'] == selected_s1_id_for_line].sort_values('s2_id')

if not df_filtered_s1.empty:
    plt.figure(figsize=(15, 6))
    sns.lineplot(data=df_filtered_s1, x='s2_id', y='mf_s1_s2', label='mf_s1_s2', alpha=0.7)
    sns.lineplot(data=df_filtered_s1, x='s2_id', y='mf_s2_s1', label='mf_s2_s1', alpha=0.7)
    # Add other metrics if desired, but keep it readable
    plt.title(f'Metric Trends for s1_id: {selected_s1_id_for_line} across its Interacting s2_ids')
    plt.xlabel('s2_id')
    plt.ylabel('Value')
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print(f"No data for selected s1_id: {selected_s1_id_for_line}")
```
